```python
# Seaborn 라이브러리
## Matplotlib 기반으로 색상 테마와 통계 기능을 추가한 패키지
## 통계 기능은 Statsmodels 패키지에 의존적
## 아나콘다 파이썬에 내장되어 있으며 만약 없다면 conda install seaborn 혹은 pip install seaborn 을 통해 설치

%matplotlib inline
import seaborn as sns

rc = {
    'axes.unicode_minus' : False,
    'font.family' : 'Malgun Gothic'
}
sns.set_style("darkgrid", rc)   # tick 스타일 지정
sns.set_palette("hls")          # 색상 팔레트 지정

import pandas as pd
df33 = pd.read_excel('sample-line-2.xlsx', index_col='날짜')
ax = df33.plot(figsize = (7, 4))
```

```python
# sns.set_style을 통한 style 지정
## whitegrid, white, darkgrid, dark 등

# 색상 팔렛트 지정
## 현재 팔렛트 획득
팔렛트 = sns.color_palette()

## 지정 팔렛트 획득
팔렛트 = sns.color_palette("dark")

## 색상 확인
sns.palplot(팔렛트)

## 변경 1 ) 디폴트 팔렛트 지정
팔렛트 = sns.color_palette("dark")
sns.set_palette(팔렛트)

## 변경 2 ) 디폴드 팔렛트 지정
sns.set_palette("dark")

# 변경 3 ) 지정 block에 대해서 팔렛트 지정
with sns.color_palette(팔렛트) :
    pass
```

```python
# 보통 색상 컨셉은 하나를 지정해서 진행하기 때문에 코드 시작하는 부분에 팔렛트 지정 후 작업을 진행한다

%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns

rc = {
    'axes.unicode_minus' : False,
    'font.family' : 'Malgun Gothic'
}
sns.set_style('whitegrid', rc)
sns.set_palette('bright')
```

```python
df44 = pd.read_excel('sample-melon.xlsx').set_index('곡일련번호')
print(df44.shape)
df44.head()
```

```python
df44['가수'].value_counts()
```

```python
plt.figure(figsize=(30, 8))     # seaborn에서는 사이즈 조절 기능이 없으므로 matplotlib 기능을 빌려쓴다.
sns.countplot(data = df44, x = "가수")
plt.xticks(rotation = 90)       # x축 값이 너무 많아 보기 힘드므로 x축 값들을 90도 회전시켜 보기 좋게 바꿔준다.
```

```python
# pandas로 countplot 흉내내기
df44.groupby('가수').size().sort_values(ascending=False)[:10]
# 혹은 df44['가수'].value_counts().sort_values(ascending=False)[:10]
```

```python
팔렛트 = sns.color_palette('husl')
series33 = df44.groupby('가수').size()
series33.plot(kind = 'bar', figsize = (30, 8), color = 팔렛트)
```
