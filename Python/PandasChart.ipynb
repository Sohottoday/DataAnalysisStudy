```python
# 항상 시작 전 설정
%matplotlib inline
from matplotlib import pyplot as plt

plt.rcParams['axes.unicode_minus'] = False  # 폰트 이슈
plt.rcParams['font.family'] = 'Malgun Gothic'   # 운영체제 맞게 쓰기

import pandas as pd

# 엑셀 파일을 읽어 올 경우
# pd.read_excel 을 위해 xlrd 설치
```

```python
line1_df = pd.read_excel("data/sample-line-1.xlsx", index_col="날짜")
print(line1_df.shape)
line1_df.head()
```

```python
#line1_df['체결가'].plot()

# plot() 위에 커서를 두고 Shift + Tab을 누르면 설명이 표시된다.
line1_df['체결가'].plot(kind = 'line')
```

```python
line2_df = pd.read_excel("data/sample-line-2.xlsx", index_col="날짜")
line2_df.head()
```

```python
line2_df.plot()     # 컬럼으로 수치데이터만 들어있으므로 Series가 차트로 그려진다.
```

```python
# Pie 타입

pie_df = pd.read_excel('data/sample-pie.xlsx', index_col="가수")

pie_df.iloc[:10].plot(kind = 'pie', y = '곡수', legend = False, figsize = (5, 5))       # 값이 너무 많으므로 iloc를 사용해 상위 10개만 지정
```

```python
# bar 타입
bar_df = pd.read_excel('data/sample-bar.xlsx', index_col='가수')
bar_df.plot(kind = 'bar', legend = False, figsize = (15, 5), rot = 0)   # rot값을 0으로 줬기 때문에 x축 이름이 돌아가 있는 상태가 아닌 원래의 상태로 돌아온다.
# legend 형식은 뭘 설정하는건지?
```

```python
bar_df['곡수'].plot(kind = 'bar', legend = False, figsize = (15, 5), rot = 0)   # 곡수를 기준으로 잡으면 Series 형식으로 변환되어 곡수를 기준으로 출력된다.( 색 변환)
```

```python
# barh 타입
barh1_df = pd.read_excel('data/sample-barh-1.xlsx', index_col='가수')
barh1_df.iloc[-10:].plot(kind = 'barh', legend = False)

barh2_df = pd.read_excel('data/sample-barh-2.xlsx', index_col='가수')
barh2_df.iloc[-10:].plot(kind = 'barh', legend = False)     # 상위 10개 출력
```

```python
# Scatter 타입
scatter_df = pd.read_excel('data/sample-scatter.xlsx')
scatter_df.plot(kind = 'scatter', x = 'S&P500 - Ratio', y = 'KOSPI200 - Ratio', figsize = (10, 10))

import seaborn as sns           # 추세선 그려보기
plt.figure(figsize = (10, 10))
sns.regplot(data = scatter_df, x = 'S&P500 - Ratio', y = 'KOSPI200 - Ratio')

# seaborn 에서는 figsize를 지원하지 않으므로 변수에 담은 뒤 plt를 사용해 사이즈를 조절한다.
```
