```python
from pykrx import stock
```

```python
df = stock.get_market_ohlcv_by_date("20200101", "20200831", "005930")
df
```

```python
print(df.iloc[df.shape[0]-1][3])                # iloc property
```

```python
cond = df["종가"] > df["시가"]
print(df[cond])
```

```python
print("상승일:", len(df[cond]))
print("영업일:", len(df))
```

```python
df["전날거래량"] = df["거래량"].shift(1)
df
```

```python
cond = df["거래량"] > df["전날거래량"]

print("전날대비 거래량 증가일수:", len(df[cond]))
print("영업일:", len(df))
```

```python
df['5일이동평균'] = df['종가'].rolling(window=5).mean()
df['20일이동평균'] = df['종가'].rolling(window=20).mean()
df['60일이동평균'] = df['종가'].rolling(window=60).mean()
df
```
