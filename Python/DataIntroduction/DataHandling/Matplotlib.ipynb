```python
# 하나의 Figure에 다수의 Axes 구조

## Figure 크기 지정하기
%matplotlib inline
from matplotlib import pyplot as plt

x = [0, 1, 2, 3]
y = [1, 4, 9, 16]

plt.figure(figsize=(2, 1))     # 인치 단위의 사이즈

plt.plot(x,y)
```

```python
plt.rc('axes', unicode_minus = False)       # - 이슈, axes라고 줬기 때문에 축에 대한 설정을 하겠다는 의미.
plt.rc('font', family = 'Malgun Gothic')        # 그래프에 한글 잘려나오는 것 방지

# plt.rcParams['axes.unicode_minus'] = False
# plt.rcParams['font.family'] = 'Malgun Gothic'     이렇게 지정해도 된다.
```

```python
# 사용 가능한 폰트 목록
import os
import pandas as pd
from matplotlib.font_manager import fontManager

font_list = []
for font in fontManager.ttflist:
    font_list.append({      # 딕셔너리 형식으로
        'name' : font.name,
        'file_name' : os.path.basename(font.fname),
        'style' : font.style
    })

df = pd.DataFrame(font_list).set_index('name')
print(df.shape)
df.head()
```

```python
# 꾸미기

%matplotlib inline
from matplotlib import pyplot as plt

x = [0, 1, 2, 3]
y = [1, 4, 9, 16]

plt.figure(figsize=(10, 5))

plt.plot(x,y)

plt.title("Line chart")
plt.xlabel("Label X")
plt.ylabel("Y 축")
plt.grid()      # 격자선 표현
```

```python
# 스타일 시트 지정

%matplotlib inline
from matplotlib import pyplot as plt

#plt.style.use('seaborn-pastel')
#plt.style.use('ggplot')
#plt.style.use('fast')
plt.style.use('seaborn-darkgrid')       # 스타일 지정, 보통 상단에서 지정한다.

x = [0, 1, 2, 3]
y = [1, 4, 9, 16]

plt.figure(figsize=(10, 5))

plt.plot(x,y)

plt.title("Line chart")
plt.xlabel("Label X")
plt.ylabel("Y 축")
plt.grid()
```

```python
# pandas 활용

%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.style.use('fast')

x = [0, 1, 2, 3]
y = [1, 4, 9, 16]

df = pd.DataFrame(y, index=x)
#df.plot()
ax = df.plot(title = "Line Plot", figsize = (8, 4), grid = True)

ax.set_xlabel("X축")
ax.set_ylabel("Y축")
```
